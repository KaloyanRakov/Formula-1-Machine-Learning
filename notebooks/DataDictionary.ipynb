# 2024 Season Data

```python
import pandas as pd
df = pd.read_csv('Formula1_2024season_raceResults.csv')
import pandas as pd
def infer_type(series):
    if pd.api.types.is_integer_dtype(series):
        return "Integer"
    elif pd.api.types.is_float_dtype(series):
        return "Float"
    elif pd.api.types.is_bool_dtype(series):
        return "Boolean"
    else:
        return "String"

data_dict = []

for col in df.columns:
    col_data = df[col]
    inferred_type = infer_type(col_data)
    required = "Yes" if col_data.isnull().sum() == 0 else "No"
    example = col_data.dropna().iloc[0] if not col_data.dropna().empty else "N/A"

    if col == "Set Fastest Lap":
        allowed_values = "Yes, No"
    elif col == "Time/Retired":
        allowed_values = 'Time format (e.g., 1:31:44.742, +25.110) or DNF'
    elif col == "Fastest Lap Time":
        allowed_values = 'Time format MM:SS.sss'
    elif inferred_type == "String":
        allowed_values = "Any string"
    elif inferred_type == "Integer":
        allowed_values = "Integer ≥ 0"
    else:
        allowed_values = "Any"

    data_dict.append([col, inferred_type, required, allowed_values, example])

data_dict_df = pd.DataFrame(data_dict, columns=["Field Name", "Data Type", "Required", "Allowed Values / Format", "Example"])
data_dict_df.head(15)
```
